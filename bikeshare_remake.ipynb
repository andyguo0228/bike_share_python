```python
import requests
import json
import pandas as pd


def jprint(obj):
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)

bike = requests.get('https://gbfs.divvybikes.com/gbfs/en/station_information.json')
jprint(bike.json())

bike_json = bike.json()
bike_data = bike_json['data']['stations']
df = pd.DataFrame(bike_data)
df.info()
```
